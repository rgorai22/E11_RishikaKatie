# Source 1 

```python
datasets = [al1, al2, al3, al4, alloy1, alloy2, alloy3, unblocked_2min]
labels = ['1 Al', '2 Al', '3 Al', '4 Al', '1 W', '2 W', '3 W', '0 Al']
```

```python
plt.figure(figsize=(12, 8))
for data, label in zip(datasets, labels):
    counts = data[0]
    plt.hist(counts, bins=20, alpha=0.5, label=f'{label}')

plt.xlabel('Counts')
plt.ylabel('Frequency')
plt.title('Source 1 - Histograms of Counts for Different Thicknesses of Aluminum Sheets')
plt.legend()
plt.show()
```

```python
means = []
std_devs = []

plt.figure(figsize=(12, 8))
for data, label in zip(datasets, labels):
    counts = data[0] 
    mean = counts.mean()
    std_dev = counts.std()
    means.append(mean)
    std_devs.append(std_dev)
    plt.hist(counts, bins=20, alpha=0.5, label=f'{label} (μ={mean:.2f}, σ={std_dev:.2f})')

plt.xlabel('Counts')
plt.ylabel('Frequency')
plt.title('Source 1 - Histograms of Counts for Different Thicknesses of Aluminum Sheets')
plt.legend()
plt.show()

for label, mean, std_dev in zip(labels, means, std_devs):
    print(f'{label}: Mean = {mean:.2f}, Standard Deviation = {std_dev:.2f}')
```

## How do the mean and standard deviation compare?

The mean count for 0 Al is significantly higher at 56.00, which is expected since there is no material absorbing the radiation. The standard deviation is also higher at 8.98, reflecting the greater variability in counts when no absorber is present.

As the thickness of the aluminum sheets increases, the mean counts decrease from 5.25 to 1.58. This indicates that more radiation is being absorbed by thicker aluminum sheets, resulting in fewer counts being detected.
The standard deviation also changes with thickness, but not in a consistent manner. It starts at 2.14 for 1 Al, increases to 2.67 for 2 Al, decreases to 1.98 for 3 Al, and then decreases further to 1.00 for 4 Al. The variation in standard deviation could be due to the different levels of absorption and the randomness in the radiation counting process.

Similar to aluminum, the mean counts decrease as the thickness of the tungsten blocks increases, from 2.92 for 1 W to 1.08 for 3 W. This is consistent with the expectation that thicker materials absorb more radiation.
The standard deviation decreases from 1.51 for 1 W to 1.00 for 3 W, indicating less variability in the counts as the thickness increases.

## How do the measured standard deviations compare to the theoretical uncertainty you would expect from counting statistics?

In counting statistics, the standard deviation is expected to be the square root of the mean. The square root of each mean compared to the measured standard deviation differ slightly indicating that there may be additional sources of variation or error in our measurements.

## What general trend do you see in the counts measured as a function of the thickness of absorber material?

As the thickness of the absorber material (aluminum sheets in this case) increases, the mean typically decreases, indicating that fewer counts are detected due to more radiation being absorbed.

```python
Al_thickness = [1, 2, 3, 4]  
Al_counts = [5.25, 3.25, 3.08, 1.58]  
Al_uncertainties = [2.14, 2.67, 1.98, 1.00]  

W_thickness = [1, 2, 3]
W_counts = [2.92, 1.25, 1.08]  
W_uncertainties = [1.51, 1.06, 1.00] 

plt.figure(figsize=(10, 6))
plt.errorbar(Al_thickness, Al_counts, yerr=Al_uncertainties, fmt='o-', label='Aluminum (Al)', capsize=5)
plt.errorbar(W_thickness, W_counts, yerr=W_uncertainties, fmt='o-', label='Tungsten (W)', capsize=5)

plt.xlabel('Thickness of Absorber Material')
plt.ylabel('Mean Counts (with uncertainties)')
plt.title('Mean Counts vs. Thickness of Absorber Material (Log Scale)')
plt.yscale('log')
plt.legend()
plt.show()
```

Based on the data, both aluminum (Al) and tungsten (W) show a decreasing trend in mean counts with increasing thickness, which is consistent with the expected behavior. 

```python
# absorption coefficient

data_Al = pd.DataFrame({
    'Sheets': [1, 2, 3, 4],
    'Mean_Counts': [5.25, 3.25, 3.08, 1.58]
})

data_W = pd.DataFrame({
    'Sheets': [1, 2, 3],
    'Mean_Counts': [2.92, 1.25, 1.08]
})

data_Al['Ln_Counts'] = np.log(data_Al['Mean_Counts'])
data_W['Ln_Counts'] = np.log(data_W['Mean_Counts'])

slope_Al, intercept_Al = np.polyfit(data_Al['Sheets'], data_Al['Ln_Counts'], 1)
absorption_coefficient_Al = -slope_Al

slope_W, intercept_W = np.polyfit(data_W['Sheets'], data_W['Ln_Counts'], 1)
absorption_coefficient_W = -slope_W

print(f"Absorption Coefficient for Aluminum (Al): {absorption_coefficient_Al}")
print(f"Absorption Coefficient for Tungsten (W): {absorption_coefficient_W}")
```

```python
Al_thickness_s2 = [1, 2, 3]
Al_counts_s2 = [26.00, 24.00, 21.83]
Al_uncertainties_s2 = [4.18, 3.69, 4.99]

W_thickness_s2 = [1, 2, 3]
W_counts_s2 = [19.58, 6.67, 4.75]
W_uncertainties_s2 = [7.76, 4.08, 3.36]

plt.figure(figsize=(10, 6))
plt.errorbar(Al_thickness_s2, Al_counts_s2, yerr=Al_uncertainties_s2, fmt='o-', label='Aluminum (Al) - Source 2', capsize=5)
plt.errorbar(W_thickness_s2, W_counts_s2, yerr=W_uncertainties_s2, fmt='o-', label='Tungsten (W) - Source 2', capsize=5)

plt.xlabel('Thickness of Absorber Material')
plt.ylabel('Mean Counts (with uncertainties)')
plt.title('Mean Counts vs. Thickness of Absorber Material (Log Scale) - Source 2')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
data_Al_s2 = pd.DataFrame({
    'Sheets': [1, 2, 3],
    'Mean_Counts': [26.00, 24.00, 21.83]
})

data_W_s2 = pd.DataFrame({
    'Sheets': [1, 2, 3],
    'Mean_Counts': [19.58, 6.67, 4.75]
})

data_Al_s2['Ln_Counts'] = np.log(data_Al_s2['Mean_Counts'])
data_W_s2['Ln_Counts'] = np.log(data_W_s2['Mean_Counts'])

slope_Al_s2, intercept_Al_s2 = np.polyfit(data_Al_s2['Sheets'], data_Al_s2['Ln_Counts'], 1)
absorption_coefficient_Al_s2 = -slope_Al_s2

slope_W_s2, intercept_W_s2 = np.polyfit(data_W_s2['Sheets'], data_W_s2['Ln_Counts'], 1)
absorption_coefficient_W_s2 = -slope_W_s2

print(f"Absorption Coefficient for Aluminum (Al) - Source 2: {absorption_coefficient_Al_s2}")
print(f"Absorption Coefficient for Tungsten (W) - Source 2: {absorption_coefficient_W_s2}")
```

The measured absorption coefficients should be the same since it is a property of the material (aluminum or tungsten) and should not depend on the source of radiation. However, the calculated values are different due to experimental errors, minor differences in sheet thicknesses, detection equipment errors.
