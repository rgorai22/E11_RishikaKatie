```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

outsideurl = "outsidedata_group2.csv" 
outsidedf = pd.read_csv(outsideurl)

insideurl = "insidedata_group2.csv"
insidedf = pd.read_csv(insideurl)

display(outsidedf)
outsidedf.describe()
display(insidedf)
insidedf.describe()
```

```python
plt.plot(outsidedf["PM25"],label='Outside')
plt.plot(insidedf["PM25"],label='Inside')
plt.legend(loc='upper right')
plt.ylabel('Counts')                       
plt.xlabel('Time [Since Initial Measurement]')
plt.title("Counts of Pm25 vs Time for Inside and Outside Data")                                  
plt.show()
```

```python
plt.plot(outsidedf["PM100"],label='Outside')
plt.plot(insidedf["PM100"],label='Inside')
plt.legend(loc='upper right')
plt.ylabel('Counts')                       
plt.xlabel('Time [Since Initial Measurement]')
plt.title("Counts of Pm100 vs Time for Inside and Outside Data")                                  
plt.show()
```

```python
plt.hist(outsidedf["PM100"],label='Outside')
plt.hist(insidedf["PM100"],label='Inside')
plt.legend(loc='upper right')
plt.ylabel('Counts')                       
plt.xlabel('Time [Since Initial Measurement]')
plt.title("Counts of Pm100 vs Time for Inside and Outside Data")                                  
plt.show()
```

```python
sigmamin25 = (np.std(insidedf["PM25"]))/np.sqrt(len(insidedf["PM25"]))
meanin25 = np.mean(insidedf["PM25"])
#sigmamout25 = (np.std(outsidedf["Pm25"]))/np.sqrt(len(outsidedf["Pm25"]))
sigmam25 = (np.std(insidedf["PM25"])+np.std(outsidedf["PM25"]))/np.sqrt(len(outsidedf["PM25"])+len(insidedf["PM25"]))
meanout25 = np.mean(outsidedf["PM25"])
meandiff25 = abs(meanin25-meanout25)
sigmam253 = 3*abs(sigmam25)
meanstd25 = meandiff25 / sigmamin25

if sigmam253 < meandiff25:
    print("Pm25 data sets are statistically different!")
else:
    print("Pm25 data sets are statistically similar, means within 3sigma of uncertainty of the mean.")

print("Inside mean standard deviations the inside and outside measurements are separated by for PM25: ", meanstd25)
print("Mean Uncertainty for PM25: ", sigmam25)

    
sigmam100 = (np.std(insidedf["PM100"])+np.std(outsidedf["PM100"]))/np.sqrt(len(outsidedf["PM100"])+len(insidedf["PM100"]))
sigmamin100 = (np.std(insidedf["PM100"]))/np.sqrt(len(insidedf["PM100"]))
meanin100 = np.mean(insidedf["PM100"])
#sigmamout100 = (np.std(outsidedf["Pm100"]))/np.sqrt(len(outsidedf["Pm100"]))
meanout100 = np.mean(outsidedf["PM100"])
meandiff100 = abs(meanin100-meanout100)
sigmam1003 = 3*abs(sigmam100)
meanstd100 = meandiff100 / sigmamin100


if sigmam1003 < meandiff100:
    print("Pm100 data sets are statistically different!")
else:
    print("Pm100 data sets are statistically similar, means within 3sigma of uncertainty of the mean.")

print("Inside mean standard deviations the inside and outside measurements are separated by for PM100: ", meanstd100)
print("Mean Uncertainty for PM100: ", sigmam100)
```
